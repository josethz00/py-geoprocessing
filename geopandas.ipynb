```python
def path_list():
    path_municipios = os.getcwd() + '/shapefiles/municipios.shp'
    path_aeroportos = os.getcwd() + '/shapefiles/aeroportos.shp'
    path_rodovias = os.getcwd() + '/shapefiles/rodovias.shp'

    return [path_municipios, path_aeroportos, path_rodovias]

def open_shapefiles():
    list_shapefiles = path_list()

    shapefiles = []
    for file in list_shapefiles:
        shapefiles.append(gpd.read_file(file))

    return shapefiles

def filter_municipios(state: str, shapes: list[str]):
    return municipios_df[municipios_df['uf'] == state]

def airports(shape, municipio_filtrado):
    gdf = gpd.sjoin(shape, municipio_filtrado, how='inner')
    total_airports = gdf.value_counts('TipoAero')
    total_pav = gdf.value_counts('pavimento')

    return gdf, total_airports, total_pav

def run(state):
    shapes = open_shapefiles()
    municipio_filter = filter_municipios(state, shapes[0])
    gdf, total_airports, pavimento = airports(shapes[1], municipio_filter.to_crs(epsg=4674))
    print(f'Estatisticas dos aeroportos do estado do {state}')
    print(total_airports)
    print(pavimento)

    gdf.plot()

    return municipio_filter
    
```

```python
path_list()
```

```python
run('SP')
```
