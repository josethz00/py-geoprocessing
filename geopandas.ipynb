```python
import geopandas as gpd
import os
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
path = os.getcwd() + '/shapefiles/municipios.shp'
municipios_df = gpd.read_file(path)
```

```python
municipios_df.plot()
```

```python
municipios_df['area'] = municipios_df['geometry'].area  # calcula a area de cada municipio, pois a coluna geometry é um poligono e a coluna de area estava vazia
```

```python
municipios_df
```

```python
municipios_rj = municipios_df.loc[municipios_df['uf'] == 'RJ']
```

```python
municipios_rj.to_file(f'{os.getcwd()}/shapefiles/municipios_rj.shp')  # salva o arquivo
```

```python
municipios_rj.plot()
```

```python
municipios_rj['area'] = municipios_rj['geometry'].area
```

```python
municipios_rj['classe'] = municipios_rj['area'].apply(lambda x: 'Grande' if x > 0.09 else 'Pequena')
```

```python
municipios_rj
```

```python
municipios_rj.plot(column='classe', legend=True)
```

```python
municipios_rj.loc[municipios_rj['classe'] == 'Grande']
```

```python
municipios_rj.loc[municipios_rj['classe'] == 'Pequena']
```

```python
municipios_rj.loc[municipios_rj['municipio'] == 'Sapucaia']
```

```python
# plot no estado e destacando o municipio de Sapucaia
ax = municipios_rj.plot(color='#f5f5f5', edgecolor='#333')
municipios_rj.loc[municipios_rj['municipio'] == 'Sapucaia'].plot(ax=ax, color='red')
```

```python
ax = municipios_rj.plot(color='#FA8072')
municipios_rj[municipios_rj['municipio'] == 'Rio De Janeiro'].plot(ax=ax, color='red')

rio_patch = mpatches.Patch(color='red', label='Capital')
rio_entorno_patch = mpatches.Patch(color='#FA8072', label='Others')
ax.legend(handles=[rio_patch, rio_entorno_patch])

plt.show()
```
