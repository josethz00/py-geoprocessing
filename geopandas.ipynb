```python
municipios_df['area'] = municipios_df['geometry'].area  # calcula a area de cada municipio, pois a coluna geometry  um poligono e a coluna de area estava vazia
```

```python
municipios_df
```

```python
municipios_rj = municipios_df.loc[municipios_df['uf'] == 'RJ']
```

```python
municipios_rj.to_file(f'{os.getcwd()}/shapefiles/municipios_rj.shp')  # salva o arquivo
```

```python
municipios_rj.plot()
```

```python
municipios_rj['area'] = municipios_rj['geometry'].area
```

```python
municipios_rj['classe'] = municipios_rj['area'].apply(lambda x: 'Grande' if x > 0.1 else 'Pequena')
```

```python
municipios_rj
```

```python
municipios_rj.plot(column='classe', legend=True)
```
