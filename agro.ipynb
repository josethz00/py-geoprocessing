```python
path = os.getcwd() + '/USINAS_CANA_BR_ATIVAS_2021/USINAS_CANA_BR_ATIVAS_2021.shp'
sugarcane_plants_br_df = gpd.read_file(path)
```

```python
path = os.getcwd() + '/shapefiles/municipios.shp'
municipios_df = gpd.read_file(path)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

# Plotting the municipalities dataframe
municipios_df.plot(ax=ax, edgecolor='none')

# Plotting the sugarcane dataframe with no fill
sugarcane_plants_br_df.boundary.plot(ax=ax, color='red', alpha=0.5, linewidth=2)

plt.title('Sugarcane mills in Brazil')
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

# Plotting the municipalities dataframe
municipios_df.plot(ax=ax, edgecolor='none', column='regiao', legend=True)

# Plotting the sugarcane dataframe with no fill
sugarcane_plants_br_df.boundary.plot(ax=ax, color='red', alpha=0.5, linewidth=2)

plt.tight_layout()
plt.title('Sugarcane mills in Brazil - by region')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7, 7))

# Plotting the municipalities dataframe
municipios_alagoas_df = municipios_df[municipios_df['uf'] == 'AL']
municipios_alagoas_df.plot(ax=ax)

# Plotting the sugarcane dataframe with no fill
alagoas_sugarplants = sugarcane_plants_br_df.loc[sugarcane_plants_br_df['UF'] == 'AL']
alagoas_sugarplants.boundary.plot(ax=ax, color='red', alpha=0.5)

plt.tight_layout()
plt.title('Sugarcane mills in Alagoas')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(9, 9))

# Plotting the municipalities dataframe
municipios_alagoas_df = municipios_df[municipios_df['uf'] == 'AL']
municipios_alagoas_df.plot(ax=ax, color='#D8D9DA')

# Plotting the sugarcane dataframe
alagoas_sugarplants = sugarcane_plants_br_df.loc[(sugarcane_plants_br_df['UF'] == 'AL') & (sugarcane_plants_br_df['UNIDADE'] == 'USINA CAETE S.A.')]

# Notice how I removed the .boundary here to fill the shape
alagoas_sugarplants.plot(ax=ax, alpha=0.5, legend=True, column='UNIDADE')

plt.tight_layout()
plt.title('Sugarcane mills in Alagoas')
plt.show()
```
