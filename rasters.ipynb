Each line of the array is a row of pixels, and each element of that line (array) represents a pixel. The value of the pixel is a number that can represent a category, a color, or a number.

[313., 301., 296., ..., 268., 271., 270.]

So:

313. is the value of the first pixel in that row.
301. is the value of the second pixel in that row.
296. is the value of the third pixel in that row.
...
268. is one of the values towards the end of the row.
271. is the second last pixel value in the row.
270. is the last pixel value in the row.

```python
path = os.getcwd() + '/shapefiles/cbers.tif'
with rasterio.open(path, 'r') as dst:
    band1 = dst.read(1)
    band2 = dst.read(2)
    band3 = dst.read(3)
    band4 = dst.read(4)

    profile = dst.profile
```

```python
import matplotlib.pyplot as plt
from rasterio.plot import show

plt.figure(figsize=(15, 15))
show(band1, cmap='gray', transform=profile['transform'])
```

```python
fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(21, 7))
show(band1, cmap='gray', ax=ax1, transform=profile['transform'], title='Band 1')
show(band2, cmap='gray', ax=ax2, transform=profile['transform'], title='Band 2')
show(band3, cmap='gray', ax=ax3, transform=profile['transform'], title='Band 3')
show(band4, cmap='gray', ax=ax4, transform=profile['transform'], title='Band 4')
```
